```python
import pandas as pd
import numpy as np
```

```python
products = ['iphone', 'ipad', 'macbook']

revenue_df = pd.DataFrame({
    'Q1': [180, 24, 12],
    'Q2': [200, 36, 18],
    'Q3': [220, 40, 20]
},
index=products)

# Create expenses dataframe
expenses_df = pd.DataFrame({
    'Q1': [80, 10, 8],
    'Q2': [90, 16, 10],
    'Q3': [100, 20, 10]
},
index=products)

```

```python
revenue_df
```

```python
expenses_df
```

```python
# Convert dataframe to numpy arrays
revenue_arr = np.array(revenue_df)
expenses_arr = np.array(expenses_df)
```

```python
revenue_arr
```

```python
expenses_arr
```

```python
profit_arr = revenue_arr - expenses_arr
```

```python
profit_arr
```

```python
revenue_df.columns
```

```python
# We can always see the matrix in form of a dataframe in the following way
pd.DataFrame(profit_arr, index=products, columns=revenue_df.columns)
```
