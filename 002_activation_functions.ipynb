```python
import math
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
def tanh(x):
    return (
        (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
        )
```

```python
tanh(-56), tanh(50), tanh(100), tanh(-100), tanh(2)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-7), relu(1), relu(6), relu(-50)
```

```python
def leaky_relu(x):
    return max(0.1*x, x)
```

```python
leaky_relu(-100), leaky_relu(8)
```
